### Tuning Hyperparameters:

```python
from sklearn.model_selection import learning_curve, train_test_split
from sklearn.ensemble import BaggingClassifier
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import BaggingClassifier
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import accuracy_score

# Define the base estimator
base_estimator = DecisionTreeClassifier() 

# Define the bagging classifier
bagging = BaggingClassifier(base_estimator=base_estimator, random_state=42)

# Define the parameter grid to search over
param_grid = {
    'base_estimator__max_depth': [2, 4, 6, 8, 10],
    'base_estimator__min_samples_split': [2, 5, 10],
    'n_estimators': [50, 100, 150, 200]
}

# Define the grid search object
grid = GridSearchCV(estimator=bagging, param_grid=param_grid, cv=5, scoring='accuracy', n_jobs=-1)

# Fit the grid search object to the training data
grid.fit(X_train, y_train)

# Get the best hyperparameters and corresponding accuracy
best_params = grid.best_params_
best_score = grid.best_score_

# Create a new bagging classifier using the best hyperparameters
best_estimator = BaggingClassifier(base_estimator=DecisionTreeClassifier(max_depth=best_params['base_estimator__max_depth'], min_samples_split=best_params['base_estimator__min_samples_split']), n_estimators=best_params['n_estimators'], random_state=42)

# Compute the learning curve for the bagging classifier using the accuracy metric
train_sizes, train_scores, test_scores = learning_curve(best_estimator, X_train, y_train, train_sizes = np.linspace(0.1, 1.0, 20), cv=5, scoring='accuracy', n_jobs=-1, shuffle=True)

n_estimators_range = np.geomspace(10, 200, 20).astype(int)

# Compute the mean and standard deviation of the training and testing scores for each value of n_estimators
train_mean = np.mean(train_scores, axis=1)
train_std = np.std(train_scores, axis=1)
test_mean = np.mean(test_scores, axis=1)
test_std = np.std(test_scores, axis=1)

# Plot the learning curve using error bars to show the variance
plt.figure(figsize=(8, 6))
plt.title('Learning Curve')
plt.xlabel('Number of Estimators')
plt.ylabel('Accuracy')
plt.grid()
plt.fill_between(n_estimators_range, train_mean - train_std, train_mean + train_std, alpha=0.1, color='r')
plt.fill_between(n_estimators_range, test_mean - test_std, test_mean + test_std, alpha=0.1, color='g')
plt.plot(n_estimators_range, train_mean, 'o-', color='r', label='Training Score')
plt.plot(n_estimators_range, test_mean, 'o-', color='g', label='Cross-Validation Score')
plt.legend(loc='best')
plt.show()

# Fit the best estimator to the training data and evaluate its performance on the validation data
best_estimator.fit(X_train, y_train)
y_pred = best_estimator.predict(X_test)
test_accuracy = accuracy_score(y_test, y_pred)

print("Best hyperparameters:", best_params)
print("Best accuracy:", best_score)
print("Validation accuracy:", test_accuracy)
```

The plateauing of the cross-validation score after approximately 100 estimators further supports this observation, indicating that increasing the number of estimators beyond this point does not lead to a significant improvement in the model's performance on the validation set. The accuracy of the training score, on the other hand, was notably higher, ranging from 0.95 to 1.0, indicating a strong fit to the training data.

However, similar to the cross-validation score, the training score also plateaued after around 125 estimators, suggesting diminishing returns in performance with increasing complexity. These findings collectively suggest that the model is overfitting the training data, as evidenced by the significant gap between the training and cross-validation scores.

To address the issue of overfitting, it is recommended to consider reducing the complexity of the model. One approach is to decrease the number of estimators or increase the regularization to prevent the model from excessively fitting the training data. Additionally, augmenting the dataset or incorporating more informative features could help improve the model's ability to learn underlying patterns and generalize to unseen data.

In summary, the learning curve analysis underscores the importance of evaluating a model's performance on a validation set or through cross-validation. By tuning the model's hyperparameters and considering appropriate strategies to combat overfitting, we can aim to achieve the best possible performance and enhance the model's ability to generalize to new data.

Based on the analysis of impurity-based feature importances, several observations can be made:

Bias towards high cardinality features: Impurity-based importances tend to favor features with high cardinality, such as numerical features. This bias can lead to numerical features being ranked higher in importance compared to categorical features. In our case, all the top 25 features selected are numerical, starting from "lag1_q95_w3" with the highest feature importance of 0.016 and "sc_std_w2" with the the importance of 0.058. This bias can result in the omission of important categorical features from the top-ranked features.

Computed on training set statistics: Impurity-based importances are computed based on the training set statistics. Therefore, they may not fully reflect the true importance of a feature in making predictions that generalize well to the test set. This limitation arises when the model has enough capacity to fit the training set well but may not generalize effectively to unseen data. It is important to consider this when interpreting impurity-based feature importances.

To overcome these limitations and gain a more comprehensive understanding of feature importance, it is recommended to utilize permutation importances. Permutation importances provide a reliable measure of feature importance by evaluating the decrease in model performance when the values of a feature are randomly shuffled. This method is not biased towards high cardinality features and can capture the importance of both numerical and categorical features.
